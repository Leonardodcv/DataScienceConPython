```python
import pandas as pd
import uuid
```

```python
data = pd.DataFrame({
    'nombre': ['Ana', 'Juan', 'Luis'],
    'email': ['ana@example.com', 'juan@example.com', 'luis@example.com'],
    'ubicacion': ['Ciudad A', 'Ciudad B', 'Ciudad C']
})

data
```

```python
id_pseudo=[]
```

```python
for n in range(len(data)):
    id_pseudo.append(str(uuid.uuid4()))
```

```python
id_pseudo
```

```python
data["id_pseudo"] = id_pseudo
data.drop("nombre", axis=1, inplace=True)
```

```python
data
```

```python
import hashlib
```

```python
def hash_data(data):
    return hashlib.sha256(data.encode()).hexdigest()
```

```python
hash_emails = []
```

```python
for email in data["email"]:
    h_email= hash_data(email)
    hash_emails.append(h_email)

data["email"] = hash_emails
data
```

```python
mis_tokens={}
```

```python
def tokenizar(dato):
    token = str(uuid.uuid4())
    mis_tokens[token]=dato
    return token
```

```python
def recuperar_datos(token):
    return mis_tokens.get(token, "Token no valido")
```

```python
dato_original = "123-456-789"
token = tokenizar(dato_original)
```

```python
print(f"Token generado: {token}")
print(f"Dato recuperado: {recuperar_datos(token)}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
```

```python
data = pd.DataFrame({
    "edad":[22,45,36,29,58],
    "genero":["F", "M", "M", "F", "M"],
    "contratado": [1,0,1,0,1]
})
data
```

```python
agrupado = data.groupby("genero")
```

```python
for nombre, grupo in agrupado:
    print(nombre, grupo)
```

```python
data_balanceada = pd.DataFrame()
```

```python
for nombre, grupo in agrupado:
    grupo_balanceado = resample(
        grupo,
        replace=True,
        n_samples=10,
        random_state=123
    )
    data_balanceada = pd.concat([data_balanceada, grupo_balanceado])
```
