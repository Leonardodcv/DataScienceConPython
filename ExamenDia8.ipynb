```python
import numpy as np
import pandas as pd
```

```python
ruta="D:\python\datos_meteorologicos.csv"
df = pd.read_csv(ruta)
df
```

```python
temperatura = df["Temperatura"].to_numpy()
precipitacion = df["Precipitación"].to_numpy()
humedad = df["Humedad"].to_numpy()

temperatura
```

```python
temp_nulo = np.isnan(temperatura)
prep_nulo = np.isnan(precipitacion)
hume_nulo = np.isnan(humedad)
temp_nulo
```

```python
#Identificar el promedio de cada array
temp_promedio = np.nanmean(temperatura)
prep_promedio = np.nanmean(precipitacion)
hume_promedio = np.nanmean(humedad)
#Reemplazar los valores nulos con el promedio del array
temperatura[temp_nulo] = temp_promedio
precipitacion[prep_nulo] = prep_promedio
humedad[hume_nulo] = hume_promedio
temperatura
```

```python
#abordaje alternativo para lograr el mismo resultado
temp_corregido = np.where(np.isnan(temperatura), temp_promedio, temperatura)
temp_corregido
```

```python
temp_promedio = np.nanmean(temperatura)
temp_promedio
```

```python
prep_total = np.sum(precipitacion)
prep_total
```

```python
hume_maxima= np.max(humedad)
hume_maxima
```

```python
#identificar dia con mayor calor
mas_calor = np.max(temperatura)
mas_calor
```

```python
#encontrar el dia mas caluroso
mas_caluroso = np.where(temperatura == mas_calor )[0][0]
mas_caluroso
```

```python
#fecha mas calurosa
fecha_mas_calurosa = df.iloc[mas_caluroso]["Fecha"]
fecha_mas_calurosa
```

```python
fecha_mas_fria = df.iloc[np.where(temperatura == np.min(temperatura))[0][0]]["Fecha"]
fecha_mas_fria
```

```python
#Crear archivo en un DataFrame con los resultados
resultados = pd.DataFrame({
    "Metrica": ["Temperatura Promedio", "Precipitacion Total", "Humedad Maxima", "Dia mas caluroso", "Dia mas frio"],
    "Valor": [temp_promedio, prep_total, hume_maxima, fecha_mas_calurosa, fecha_mas_fria]
})
resultados.to_csv("D:\python\Resltado de analisis meteorologico.csv", index=False)
```
