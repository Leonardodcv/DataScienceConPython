```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.DataFrame()
df["Edad"]=[25, 30, 35, 40, 45]
df["Salario"]=[3000, 3500, 4000, 4500, 5000]

plt.scatter(
    df["Edad"],
    df["Salario"],
    color="blue",
    marker="o"
)

modelo=linear_model.LinearRegression()
X=df[["Edad"]]
y=df["Salario"]
modelo.fit(X,y)

plt.title("Relación entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.grid(True)

plt.scatter(
    df["Edad"],
    df["Salario"],
)
plt.plot(
    df["Edad"],
    modelo.predict(X)
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from iris_info import iris_dict
from sklearn.preprocessing import LabelEncoder

# Convertir el diccionario a DataFrame
iris_df = pd.DataFrame(iris_dict)

X=iris_df.drop("species", axis=1)

le=LabelEncoder()
especies=le.fit_transform(iris_df["species"])
y= especies

plt.scatter(X["sepal_length"], y, c=y)

plt.colorbar(label="Especies codificadas")
plt.title("Relación entre la longitud del sépalo y las especies codificadas")
plt.xlabel("Longitud del Sépalo (cm)")
plt.ylabel("Especies codificadas")
plt.grid(True)
```

```python
from iris_info import iris_dict
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

#___________IMPORTA AQUI LAS LIBRERIAS QUE REQUIERAS____________
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Convertir el diccionario que contiene los datos del dataset iris en un DataFrame de pandas para facilitar su manipulación.

iris_df = pd.DataFrame(iris_dict)

# Preparar el DataFrame para modelado, eliminando la columna 'species' y usando el resto como características (X).

X = iris_df.drop("species", axis=1)

# Inicializar el codificador de etiquetas y transformar las etiquetas categóricas de especies en valores numéricos.

le = LabelEncoder()

especies = le.fit_transform(iris_df["species"])

y = especies

#___________COMPLETA EL CÓDIGO CON LOS SIGUIENTES PASOS____________

# División de datos en conjunto de entrenamiento y de prueba
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X,y, train_size=0.8, random_state=42)
# Creación y entrenamiento del modelo de árbol de decisión

arbol=DecisionTreeClassifier()
arbol.fit(X_entrena, y_entrena)
# Visualización del árbol de decisión
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol)








```

```python
plt.figure(figsize=(20,20))
plot_tree(
    decision_tree=arbol,
    filled=True
)
```

```python
plt.figure(figsize=(20,20))
plot_tree(
    decision_tree=arbol,
    class_names=["setosa","versicolor","virginica"],
    filled=True
)
```

```python
plt.figure(figsize=(20,20))
plot_tree(
    decision_tree=arbol,
    class_names=["setosa","versicolor","virginica"],
    feature_names=df.columns.to_list(),
    filled=True
)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
escala = MinMaxScaler(feature_range=(0,1))
normalizado = escala.fit_transform(df)
df_normalizado = pd.DataFrame(data = normalizado, columns=df.columns)
df_normalizado.head()
```

```python
X = df_normalizado.drop("Clase", axis=1)
```

```python
y=df_normalizado["Clase"]
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.70, random_state=42)
```

```python
modelos=[
    ("Regresion Logistica", LogisticRegression()),
    ("Arbol de Decision", DecisionTreeClassifier()),
    ("Bosque aleatorio", RandomForestClassifier())
]
```

```python
for nombre, modelo in modelos:
    modelo.fit(X_entrena, y_entrena)
    puntaje = modelo.score(X_prueba, y_prueba)
    print(f'{nombre}:{puntaje:.6f}')
```
