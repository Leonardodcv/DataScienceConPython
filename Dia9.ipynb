```python
import matplotlib.pyplot as plt
 
# Definir los datos
y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
x = range(1, 6)  # Tiempos del 1 al 5
 
# Crear el grfico apilado
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, colors=['red', 'blue', 'green'], labels=['Serie y1', 'Serie y2', 'Serie y3'])
 
# Aadir leyenda
ax.legend(loc='upper left')
 
# Aadir ttulos y etiquetas
ax.set_title('Distribucin acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados')
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0)

data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)

data = [data1, data2, data3]
 
# Crear el violin plot
fig, ax = plt.subplots()
ax.violinplot(data)
 
# Establecer las etiquetas del eje x
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])
 
# Aadir ttulo al grfico
ax.set_title('Comparacin de distribuciones en Violin Plot')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ruta="D:\python\Datos+Meteorolgicos_Arg_2023.csv"
df=pd.read_csv(ruta)
df.head()
```

```python
df["Fecha"]= pd.to_datetime(df["Fecha"], format="%d/%m/%Y")
df.head()
```

```python
lista_ciudades= []

for ciudad in df["Ciudad"]:
    if ciudad not in lista_ciudades:
        lista_ciudades.append(ciudad)

lista_ciudades
```

```python
dict_meses={
    1:"Enero",
    2:"Febrero",
    3:"Marzo",
    4:"Abril",
    5:"Mayo",
    6:"Junio",
    7:"Julio",
    8:"Agosto",
    9:"Septiembre",
    10:"Octubre",
    11:"Noviembre",
    12:"Diciembre"
}
dict_meses
```

```python
def consultar_temperaturas():
    while True:
        #solicitar la ciudad, mes, inputs, dataframe, graficacin de temperaturas y preguntar si se debe repetir el sistema.
        print("\nCiudades disponibles: ", lista_ciudades)
        ciudad_elegida = input("Elige la ciudad de la lista").title()
    
        mes_elegido=int(input("Elije un mes usando un numero: "))
        print("ciudad_elegida:",ciudad_elegida)
        print("mes_elegido:",mes_elegido)
        
        if ciudad_elegida not in lista_ciudades:
            print("Ciudad invalidos. Intentelo nuevamente")
        if mes_elegido not in range(1, 13):
            print("Mes invalidos. Intentelo nuevamente")
            #continue

        datos_ciudad_mes = df[(df["Ciudad"]==ciudad_elegida) & (df["Fecha"].dt.month == mes_elegido)]

        plt.figure(figsize=(10,6))
        plt.plot(datos_ciudad_mes["Fecha"], datos_ciudad_mes["Temperatura Maxima"], label="Maxima", color="red")
        plt.plot(datos_ciudad_mes["Fecha"], datos_ciudad_mes["Temperatura Minima"], label="Minima", color="blue")
        plt.title(f"Temperaturas en {ciudad_elegida} durante el mes de  {dict_meses[mes_elegido]}")
        plt.xlabel("Fecha")
        plt.ylabel("Temperaturas")
        plt.legend()
        plt.xticks(rotation=45)
        plt.show()

        otra_consulta =input("Deseas continuar consultando? (s/n):")
        if otra_consulta.lower() != "s":
            break
        







        
```

```python
consultar_temperaturas()
```
