```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
diamantes=sns.load_dataset("diamonds")
```

```python
diamantes.head(10)
```

```python
diamantes.describe()
```

```python
sns.set(style="whitegrid")

g=sns.relplot(
    data=diamantes,
    x="carat",
    y="price",
    kind="scatter",
    height=6,
    aspect=1.5
)

g.fig.suptitle(
    "Relacion entre el precio y el peso en quilates de los diamantes",
    va="baseline",
    ha="center"
)

g.set_axis_labels(
    "Peso en quilates", 
    "Precio"
)
```

```python
#creacion de la vizualizacion de la distribucion del precio
g= sns.displot(
    data=diamantes, 
    x="price",
    kind="hist",
    kde=True,
    color="blue",
    height=4,
    aspect=2
)

g.fig.suptitle(
    "Distribucion del precio de los diamantes",
    va="baseline",
    ha="center"
)

g.set_axis_labels("Precio", "Frecuencia")
```

```python
#Creacioon de la visualizacion de la distribucion del peso en quilates
g = sns.histplot(
    data=diamantes,
    x="carat",
    kde=True,
    color="green"
)
g.set_title("Distribucin del peso en quilates")
g.set(xlabel="Peso en quilates(Carat)", ylabel="Frecuencia")
```

```python
#creacion de ungrafico catplot de tipo Box
sns.set(style="whitegrid")
g = sns.catplot(
    data=diamantes,
    kind="box",
    x="cut",
    y="price",
    height=5,
    aspect=2,
    order=["Fair","Good","Very Good","Premium","Ideal"]
)

# Creacion de titulo:
g.fig.suptitle(
    "Comparacion del precio de los diamantes por calidad del corte",
    va="baseline",
    ha="center"
)
# Etiquetas de los ejes X Y
g.set_axis_labels(
    "Calidad del corte",
    "Precio($)"
)
```

```python
# uso de joinplot para visualizarla relacion entre la profundidad de los diamantes y su precio en un 
# grafico de densidad kernek

g=sns.jointplot(
    data=diamantes,
    x="depth",
    y="price",
    kind="kde",
    fill=True,
    space=0,
    color="purple"
)

g.set_axis_labels(
    "Profundidad del diamante(%)", 
    "Precio"
)

g.fig.suptitle(
    "Relacion  entre la profundidad y el precio de los diamantes",
    va="baseline",
    ha="center",
    fontsize=16
);
```

```python
# Creacion de pairplot para mostrar las relaciones entre todas las variables
# Seaborn ignora las variables no numericas.

g=sns.pairplot(
    diamantes,
    corner=True
)

g.fig.suptitle(
    "Relacion entre las variables numericas en la informacion de los diamantes",
    va="baseline",
    ha="center",
    fontsize=16
);
```
