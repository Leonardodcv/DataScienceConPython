```python
grafico= sns.PairGrid(
    pinguinos,
    hue="species",
    corner=True
)
grafico.map_lower(
    sns.kdeplot,
    hue=None,
    level=5,
    color=".2"
)
grafico.map_lower(
    sns.scatterplot,
    marker="+"
);
```

```python
grafico= sns.PairGrid(
    pinguinos,
    hue="species",
    corner=True
)
grafico.map_lower(
    sns.kdeplot,
    hue=None,
    level=5,
    color=".2"
)
grafico.map_lower(
    sns.scatterplot,
    marker="+"
);
grafico.map_diag(
    sns.histplot,
    element="step",
    linewidth=0,
    kde=True
)
```

```python
grafico= sns.PairGrid(
    pinguinos,
    hue="species",
    corner=True
)
grafico.map_lower(
    sns.kdeplot,
    hue=None,
    level=5,
    color=".2"
)
grafico.map_lower(
    sns.scatterplot,
    marker="+"
);
grafico.map_diag(
    sns.histplot,
    element="step",
    linewidth=0,
    kde=True
)
grafico.add_legend(frameon=True);
```

```python
grafico= sns.PairGrid(
    pinguinos,
    hue="species",
    corner=True
)
grafico.map_lower(
    sns.kdeplot,
    hue=None,
    level=5,
    color=".2"
)
grafico.map_lower(
    sns.scatterplot,
    marker="+"
);
grafico.map_diag(
    sns.histplot,
    element="step",
    linewidth=0,
    kde=True
)
grafico.add_legend(frameon=True);
grafico.legend.set_bbox_to_anchor((.61,.6))
```

```python
sns.set_theme(style="dark")

sns.relplot(
    data=pinguinos,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    palette="crest",
    marker="x",
    s=100
)
```

```python
g=sns.relplot(
    data=pinguinos,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    palette="crest",
    marker="x",
    s=100
)

g.set_axis_labels("largo del pico en mm", "Grosor del picos en mm", labelpad=10);
```

```python
g=sns.relplot(
    data=pinguinos,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    palette="crest",
    marker="x",
    s=100
)

g.set_axis_labels(
    "largo del pico en mm", 
    "Grosor del picos en mm", 
    labelpad=10
);

g.legend.set_title("Masa\nCorporal (g)")
```

```python
g=sns.relplot(
    data=pinguinos,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    palette="crest",
    marker="x",
    s=100
)

g.set_axis_labels(
    "largo del pico en mm", 
    "Grosor del picos en mm", 
    labelpad=10
);

g.legend.set_title("Masa\nCorporal (g)")
g.figure.set_size_inches(6.5, 4.5)
```

```python
g=sns.relplot(
    data=pinguinos,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="body_mass_g",
    palette="crest",
    marker="x",
    s=100
)

g.set_axis_labels(
    "largo del pico en mm", 
    "Grosor del picos en mm", 
    labelpad=10
);

g.legend.set_title("Masa\nCorporal (g)")
g.figure.set_size_inches(6.5, 4.5)
# sirve para cambiar el porcentaje de margen que tiene el grafico y en este caso es 
# 15 porciento
g.ax.margins(0.15)
```
