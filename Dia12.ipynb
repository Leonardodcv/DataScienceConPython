```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
data = np.random.rand(100,2)
data
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data[:,0], data[:, 1], c=etiquetas, cmap="viridis", marker="o")
plt.scatter(centroides[:,0], centroides[:,1], c="red", marker="x", s=200, label="Centroides")
plt.title("Vizualizacion de K-Means Clustering")
plt.legend();
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generar datos aleatorios
np.random.seed(42)
datos = np.random.rand(100, 2)

# Definir el modelo KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

# Ajustar el modelo a los datos
kmeans.fit(datos)

# Obtener centroides y etiquetas
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_

plt.figure(figsize=(10,6))
plt.scatter(datos[:,0], datos[:,1], c=etiquetas, cmap="viridis", marker="o")
plt.scatter(centroides[:,0], centroides[:,1], c="red", marker="x", s=200, label="Centroides")
plt.title("K-Means Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend();
```

```python
iris=load_iris()
X=iris.data
```

```python
X_centrado = X-np.mean(X, axis=0)
```

```python
U, sigma, VT = np.linalg.svd(X_centrado)
```

```python
k=2
# Tranforma los datos de n dimensiones al valor de k con el objetivo de facilitar la deteccion
# de patrones o de agrupaciones en los datos
X_transformado = U[:,:k] * sigma[:k]
```

```python
especies=["setosa", "versicolor","virginica"]
```

```python
plt.figure(figsize=(8,6))
for i in range(3):
    plt.scatter(
        X_transformado[iris.target == i, 0],
        X_transformado[iris.target == i, 1],
        label=especies[i]
    )
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend()
plt.title("Dataset iris transformado por SVD")
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
```

```python
digitos = load_digits()
digitos
```

```python
digitos["data"][0]
```

```python
digitos["data"][0].reshape(8, 8)
```

```python
plt.imshow(digitos["data"][0].reshape(8,8))
```

```python
plt.imshow(digitos["data"][2].reshape(8,8))
```

```python
plt.imshow(digitos.images[0]);
```

```python
X = digitos.data
```

```python
X = X/16.0
```

```python
X_entrena, X_prueba = train_test_split(X, test_size=0.20, random_state=42)
```

```python
imagen_entrada = Input(shape=(64, ))
```

```python
# Una capa de neuronas densa modifica el dato de entrada, en este caso la imagen de entrada,
# cada neurona, de las 32 usadas, lo modifica lo modifica en base a sus propias reglas. Y la
# salida de la capa se puede hacer en base a todas las neuronas.
# Las reglas pueden cambiar en base.
codificado = Dense(32, activation="relu")(imagen_entrada)
```

```python
# se usa 64 ya que es el numero de entreda original(es decir 64 pixels en una matriz de 8X8
# sigmoid convierte los valores de entrada entre 0 y 1

decodificado = Dense(64, activation="sigmoid") (codificado)
```

```python
autoencoder=Model(imagen_entrada, decodificado)
```

```python
# loss mide que tambien puede recuperar la entrada original de los datos desde la entrada comprimida
# ademas adam recalcula los pesos en base a la diferencia de las entradas y salidas de la información
#
autoencoder.compile(
    optimizer="adam", 
    loss="binary_crossentropy"
)
```

```python
# entrenamiento del metodo
autoencoder.fit(
    X_entrena, # se le pasa los mismos datos de entrada y salida ya que el algoritmo deconstruye y 
    X_entrena, # y luego recontruye los mismos datos
    epochs=100, # epocas es el numero de veces que se repasara los datos de entrenamiento
    batch_size=256, #tamño del lote: indica el numero de datos de entrada que se tomaran al mismo tiempo
    shuffle=True, #Mezcla los datos antes de cada epoca
    validation_data= (X_prueba,X_prueba)
)
```

```python
for i in range(10):
    plt.subplot(2, 10, i+1)
    plt.imshow(X_prueba[i].reshape(8,8))
    plt.subplot(2,10,i+1+10)
    plt.imshow(autoencoder.predict(X_prueba)[i].reshape(8,8))
```

```python
#########################
# Clustering jerarquico #
#########################
# proporciona informacion acerca de la información 
```

```python
from sklearn.datasets import load_iris
import matplotlib as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
# ward se utiliza para minimizar la varianza al crear cada cluster de datos
linked= linkage(X, "ward")
```

```python
plt.figure(figsize=(10,7))
dendrogram(
    linked, # Contiene la informacion 
    orientation="top", # guarda la distancia en longitud de los diferentes clusters, ademas dibuja el arbol de arriba hacia abajo
    labels=[iris.target_names[i] for i in iris.target],  # etiquetas con nombres y como se compilan en cada paso
    distance_sort="descending", # muestra como se iran mostrando las hojas
    show_leaf_counts=True # añade el numero de clusters junto con las hojas
)
plt.title("Dendrograma de clustering Jerárquico")
plt.show()
```
